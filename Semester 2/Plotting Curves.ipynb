```python
from numpy import pi, cos, arange
import matplotlib.pyplot as plt
a=2
angle=arange(0,(2*pi),0.01)
for theta in angle:
    r=a*(1+cos(theta))
    plt.polar(theta,r,'r.')
plt.show()
```

```python
from sympy import symbols,sin,cos,pi
from sympy.plotting import plot_parametric
theta=symbols('theta')
a=3
plot_parametric(a*(theta+sin(theta)),a*(1+cos(theta)),(theta,-pi,pi))
```
